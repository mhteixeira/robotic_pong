```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colnames = [
    'is_ball_detected',
    't_0',
    't_frame',
    't_distortions',
    't_ball_detection',
    't_prediction',
    't_moving_motor',
    't_plot'
]
```

```python
df_w = pd.read_csv('./main_opencv_plot_13_11.txt', header=None, names=colnames)
for i in reversed(range(2, len(colnames))):
    df_w[colnames[i]] = df_w[colnames[i]] - df_w[colnames[i-1]]
df_w['t_total'] = df_w.iloc[:, 2:].sum(axis=1)
df_w.drop('t_0', inplace=True, axis=1)
df_w['is_ball_detected'] = df_w['is_ball_detected']*0.1
```

```python
df_wo = pd.read_csv('./main_normal_plot_13_11.txt', header=None, names=colnames)
for i in reversed(range(2, len(colnames))):
    df_wo[colnames[i]] = df_wo[colnames[i]] - df_wo[colnames[i-1]]
df_wo['t_total'] = df_wo.iloc[:, 2:].sum(axis=1)
df_wo.drop('t_0', inplace=True, axis=1)
df_wo['is_ball_detected'] = df_wo['is_ball_detected']*0.1
```

```python
df_no = pd.read_csv('./main_no_plot_13_11.txt', header=None, names=colnames)
for i in reversed(range(2, len(colnames))):
    df_no[colnames[i]] = df_no[colnames[i]] - df_no[colnames[i-1]]
df_no['t_total'] = df_no.iloc[:, 2:].sum(axis=1)
df_no.drop('t_0', inplace=True, axis=1)
df_no['is_ball_detected'] = df_no['is_ball_detected']*0.1
```

```python
plt.figure(figsize=(16,4))
plt.subplot(132)
plt.plot(df_w)
plt.ylim([0, 0.15])
plt.ylabel("Time (s)")
plt.xlabel("Frame count")
plt.legend(df_w.columns)
plt.title("OpenCV")
plt.subplot(131)
plt.plot(df_wo)
plt.ylim([0, 0.15])
plt.ylabel("Time (s)")
plt.xlabel("Frame count")
plt.title("Matplotlib")
plt.legend(df_wo.columns)
plt.subplot(133)
plt.plot(df_no)
plt.ylim([0, 0.15])
plt.ylabel("Time (s)")
plt.xlabel("Frame count")
plt.title("No plot")
plt.legend(df_wo.columns)
plt.suptitle("Different plotting methods")
```

```python
plt.figure(figsize=(12,4))
plt.subplot(122)
plt.plot(df_w['t_frame'])
# plt.legend(df_w.columns)
plt.ylim([0, 0.07])
plt.title("OpenCV")
plt.subplot(121)
plt.plot(df_wo['t_frame'])
plt.title("Matplotlib")
plt.ylim([0, 0.07])
# plt.legend(df_wo.columns)
plt.suptitle("Time difference between frame capturing time")
```
